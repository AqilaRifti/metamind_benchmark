## Overview

MetaMind evaluates four key dimensions of metacognition:

1. **Confidence Calibration**: Alignment between stated confidence and actual accuracy
2. **Error Detection & Self-Correction**: Ability to recognize and fix own mistakes
3. **Knowledge Boundary Detection**: Appropriate refusal on unanswerable questions
4. **Strategic Self-Monitoring**: Step-by-step reasoning monitoring

Let's explore each task with examples.

```python
# Install kaggle_benchmarks if needed
# !pip install kaggle_benchmarks

import kaggle_benchmarks as kbench
import json
import matplotlib.pyplot as plt
import numpy as np

print("✓ kaggle_benchmarks imported successfully")
```

## Task 1: Confidence Calibration

This task measures how well a model's confidence matches its actual accuracy.

```python
from task_confidence_calibration import CONFIDENCE_CALIBRATION_DATASET, run_confidence_calibration_benchmark

# Show sample questions
print("Sample Questions from Confidence Calibration Dataset:")
print("="*60)
for i, q in enumerate(CONFIDENCE_CALIBRATION_DATASET[:3]):
    print(f"\n{i+1}. [{q['difficulty'].upper()}] {q['domain']}")
    print(f"   Q: {q['question']}")
    print(f"   A: {q['answer']}")
```

```python
# Run confidence calibration task
llm = kbench.llm
calibration_results = run_confidence_calibration_benchmark(llm)

print("\nConfidence Calibration Results:")
print("="*60)
print(f"Expected Calibration Error (ECE): {calibration_results['expected_calibration_error']:.4f}")
print(f"Overall Accuracy: {calibration_results['overall_accuracy']:.2%}")
print(f"Average Confidence: {calibration_results['average_confidence']:.2%}")
print(f"Overconfidence Score: {calibration_results['overconfidence_score']:+.4f}")

if calibration_results['overconfidence_score'] > 0.1:
    print("\n⚠️  Model shows significant OVERCONFIDENCE")
elif calibration_results['overconfidence_score'] < -0.1:
    print("\n⚠️  Model shows UNDERCONFIDENCE")
else:
    print("\n✓ Model shows good calibration")
```

## Task 3: Knowledge Boundary Detection

Tests whether models appropriately refuse unanswerable questions.

```python
from task_knowledge_boundaries import KNOWLEDGE_BOUNDARY_DATASET, run_knowledge_boundary_benchmark

# Show question categories
categories = {}
for q in KNOWLEDGE_BOUNDARY_DATASET:
    cat = q['category']
    categories[cat] = categories.get(cat, 0) + 1

print("Knowledge Boundary Question Categories:")
print("="*60)
for cat, count in categories.items():
    print(f"  {cat}: {count} questions")
```

```python
# Run knowledge boundary benchmark
boundary_results = run_knowledge_boundary_benchmark(llm)

print("\nKnowledge Boundary Detection Results:")
print("="*60)
print(f"Overall Accuracy: {boundary_results['overall_accuracy']:.2%}")
print(f"Answerable Questions (should answer): {boundary_results['answerable_accuracy']:.2%}")
print(f"Boundary Recognition (should refuse): {boundary_results['boundary_recognition_accuracy']:.2%}")
print(f"Hallucination Rate: {boundary_results['hallucination_rate']:.2%}")
print(f"Over-Refusal Rate: {boundary_results['over_refusal_rate']:.2%}")

print("\nCategory Breakdown:")
for cat, metrics in boundary_results['category_breakdown'].items():
    print(f"  {cat}: {metrics['accuracy']:.2%} ({metrics['correct_responses']}/{metrics['total']})")
```

## Complete MetaMind Benchmark

Now let's run the complete benchmark and generate the cognitive profile.

```python
from benchmark import metamind_benchmark

# Run complete benchmark
full_results = metamind_benchmark(llm)

print("\n" + "="*60)
print("METAMIND COGNITIVE PROFILE")
print("="*60)
print(f"\nOverall Metacognition Score: {full_results['overall_metacognition_score']:.4f}")

print("\nComponent Scores:")
for component, score in full_results['component_scores'].items():
    bar = "█" * int(score * 20) + "░" * (20 - int(score * 20))
    print(f"  {component:30s} {bar} {score:.2f}")

print(f"\nInterpretation:")
print(f"  {full_results['interpretation']}")
```

## Visualization

Let's create a radar chart of the metacognitive profile.

```python
# Create radar chart
categories = list(full_results['component_scores'].keys())
scores = list(full_results['component_scores'].values())

# Close the polygon
angles = np.linspace(0, 2 * np.pi, len(categories), endpoint=False).tolist()
scores_plot = scores + [scores[0]]
angles += angles[:1]

fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(projection='polar'))
ax.plot(angles, scores_plot, 'o-', linewidth=2, color='#2563eb')
ax.fill(angles, scores_plot, alpha=0.25, color='#2563eb')
ax.set_xticks(angles[:-1])
ax.set_xticklabels([c.replace('_', '\n') for c in categories], size=10)
ax.set_ylim(0, 1)
ax.set_title('MetaMind Cognitive Profile', size=14, fontweight='bold', pad=20)
ax.grid(True)

plt.tight_layout()
plt.show()

print(f"Overall Score: {full_results['overall_metacognition_score']:.2%}")
```

## Key Insights

### What This Benchmark Reveals

1. **Calibration Quality**: The ECE metric shows how much a model's confidence deviates from its accuracy. Lower is better.

2. **Error Awareness**: The error detection rate reveals blind spots. Models that can't recognize their own errors cannot self-improve.

3. **Hallucination Tendency**: High hallucination rates indicate models answer questions they shouldn't, a critical safety concern.

4. **Self-Monitoring Utility**: The correlation between explicit monitoring and accuracy validates that metacognition improves performance.

### Comparison to Standard Benchmarks

| Aspect | Standard Benchmarks | MetaMind |

| Measures | What models know | How well they know it |
| Output | Accuracy score | Cognitive profile |

| Use Case | Capability ranking | Deployment readiness |
